### IMPORTING NECESSARY LIBRARIES

```python
import sqlite3
import csv
import pandas as pd
import numpy as np
```

#### creating cars database

```python
con = sqlite3.connect("cartest.db")
cur = con.cursor()
```

### IMPORT DATA-SET TO BE USED

```python
data = pd.read_csv('Automobile_miles_per_gallon.csv')
data.head()
```

###  CREATING the train and test data set

```python
from sklearn.model_selection import train_test_split
#split the data into train and test set
train,test = train_test_split(data, test_size=0.20, random_state=20)
#save the data
train.to_csv('train.csv',index=False)
test.to_csv('test.csv',index=False)
```

#### Creating the table 

```python
table = """CREATE TABLE autotest
(mpg REAL, cyl INTEGER, displREAL, hp INTEGER,weight INTEGER, accel REAL, yr REAL, Origin TEXT, name TEXT)"""
```

```python
cur.execute(table)
```

```python
file = open("test.csv")
rows = csv.reader(file)
cur.executemany("INSERT INTO  autotest VALUES (?, ?, ?, ?, ?, ?, ?, ?, ? )", rows)
```
