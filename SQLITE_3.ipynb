### IMPORTING NECESSARY LIBRARIES

```python
import sqlite3
import csv
con = sqlite3.connect("cars.db")
cur = con.cursor()
```

#### Creating the table 

```python
tab = """CREATE TABLE auto
(mpg REAL, cyl INTEGER, displREAL, hp INTEGER,weight INTEGER, accel REAL, yr REAL, Origin TEXT, name TEXT)"""
```

```python
cur.execute(tab)
```

```python
file = open("Automobile_miles_per_gallon.csv")
rows = csv.reader(file)
cur.executemany("INSERT INTO  auto VALUES (?, ?, ?, ?, ?, ?, ?, ?, ? )", rows)
```

#### LETS QUERY THE DATA SET

 FETCHING THE ENTIRE SET FROM THE DATA SET

```python
cur.execute("SELECT * FROM auto")
print(cur.fetchall())
```

LET'S PRINT THE FIRST ROW

```python
cur.execute("SELECT * FROM auto")
print(cur.fetchone())
```

LET'S FETCH 4 ROWS

```python
cur.execute("SELECT * FROM auto")
print(cur.fetchmany(4))
```
